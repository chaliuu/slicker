```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

accelerations = pd.read_csv('../accelerations.csv')
# print(accelerations['xAcc'], accelerations['yAcc'])
# accelerations.head()
axis = np.linspace(0.0, 440.0, num=440)
# print(axis)
# print(accelerations['xAcc'][:-1])

plt.plot(axis, accelerations['xAcc'][:-1], 'r-', markersize=2, label = 'X - Acceleration')
plt.plot(axis, accelerations['yAcc'][:-1], 'b-', markersize=1, label = 'Y - Acceleration')
plt.title('Y / X - Acceleration vs. Time')
plt.xlabel('Time (ms)')
plt.ylabel('Acceleration (m/s^2)')
plt.legend(loc='lower right')
plt.show()
```
